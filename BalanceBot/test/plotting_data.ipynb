```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [10, 10]
```

```python
def plot_speedController(filename,kp1,kp2):
    source=np.loadtxt(filename,delimiter=',',dtype=float)
    time = np.linspace(0,len(source)*.1,len(source))
    fig, (ax1, ax2) = plt.subplots(nrows=2, sharex=True)
    ax1.set_title("Summing P Controller for Speed of Both Motors K_p={},{}".format(kp1,kp2))
    ax1.plot(time,source[:,0],label="Setpoint")
    ax1.plot(time,source[:,3],label="Motor 1")
    ax1.plot(time,source[:,6],label="Motor 2")
    plt.xlabel("Time [s]")
    ax1.set_ylabel("Speed [RPM]")
    ax1.legend()
    # plt.subplot(212,sharex=True)
    ax2.plot(time,source[:,1],label="Motor 1 Output")
    ax2.plot(time,source[:,2],label="Motor 1 Response")
    ax2.plot(time,source[:,4],label="Motor 2 Output")
    ax2.plot(time,source[:,5],label="Motor 2 Response")
    ax1.set_ylabel("Motor Strength [%]")
    ax2.legend()
    plt.figure()
    # plt.title("Summing P Controller for Speed of Both Motors")
```

```python
def plot_balanceController(filename,kp1,kp2,kd1,kd2=0):
    source=np.loadtxt(filename,delimiter=',',dtype=float)
    time = np.linspace(0,len(source)*.1,len(source))
    if kd2==0:
        plt.title("Full Controller K_p={},{} K_d={}".format(kp1,kp2,kd1))
    else:
        plt.title("Full Controller K_p={},{} K_d={},{}".format(kp1,kp2,kd1,kd2))
    plt.plot(time,source[:,0],label="Error [deg]")
    plt.plot(time,source[:,1],label="Speed Setpoint [RPM]")
    plt.plot(time,source[:,2],label="Motor 1 Speed [RPM]")
    plt.plot(time,source[:,3],label="Motor 2 Speed [RPM]")
    plt.xlabel("Time [s]")
    # plt.set_ylabel("Speed [RPM]")
    plt.legend()
    plt.figure()
```

```python
plot_balanceController("balanceData_P10_8_D8_4.csv",10,8,8,4)
plot_balanceController("balanceData_P8_5_D8_1p5.csv",8,5,8,1.5)
plot_balanceController("balanceData_P5_2_D1_p5.csv",5,2,1,.5)
plot_balanceController("balanceData_P1p5_p75_Dp5.csv",1.5,.75,.5)
plot_balanceController("balanceData_Pp75_p5_Dp5_p05.csv",.75,.5,.5,.05)
plot_balanceController("balanceData_P1p5_p75_Dp5_p05.csv",1.5,.75,.5,.05)
plot_balanceController("balanceData_P2_1p5_Dp75_p25.csv",2,1.5,.75,.25)
```
